## Pandas

```python
import pandas as pd
import numpy as np
```

### Dataset cleaning

```python
df=pd.read_csv('movies.csv')
```

```python
df.timestamp=pd.to_datetime(df.timestamp)
df.release_date=pd.to_datetime(df.release_date)
```

```python
df.rename(columns = {'Western\\':'western'}, inplace = True)
df.western = df.western.replace(to_replace= r'\\', value= '', regex=True)
df.western=df.western.replace(to_replace= r'}', value= '', regex=True)
df.western=df.western.astype(str).astype(int)
```

```python
df=df.rename(str.lower, axis=1) #axis='columns'
```
