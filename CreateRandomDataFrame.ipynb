### Create Random Data Frame

Indexes - lowercase alphabet  
Columns - short month names  
Values - random integer from '- 100' to '100'

Pandas  
Numpy  
Calendar

```python
import pandas as pd
import numpy as np
import string
import calendar
```

```python
pd.DataFrame(np.random.randint(-100,100,(6,10)),
               index=[i for i in string.ascii_lowercase[:6]],
               columns=[calendar.month_abbr[month_idx] for month_idx in range(10)])
```
